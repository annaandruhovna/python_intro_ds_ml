```python
import pandas as pd
import numpy as np
```

```python
students_perfomance = pd.read_csv('https://stepik.org/media/attachments/course/4852/StudentsPerformance.csv')
```

```python
#Переименование переменных в датафрейме
students_perfomance = students_perfomance \
    .rename(columns = 
            {'parental level of education' : 'parental_level_of_education',
             'math score' : 'math_score',
             'reading score' : 'reading_score',
             'writing score' : 'writing_score'
            })
```

```python
#Создание новой колонки в датафрейме
students_perfomance['total_score'] = students_perfomance['math_score']+students_perfomance['reading_score']+students_perfomance['writing_score']
```

```python
students_perfomance.head()
```

```python
#Создание нескольких новых колонок в датафрейме
students_perfomance = students_perfomance.assign(total_score_log = np.log(students_perfomance['total_score']))
```
