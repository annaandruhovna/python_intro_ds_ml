```python
import pandas as pd
import numpy as np
```

```python
#Аспирант Ростислав изучает метаболом водорослей и получил такую табличку. 
#В ней он записал вид каждой водоросли, её род (группа, объединяющая близкие виды), 
#группа (ещё одно объединение водорослей в крупные фракции) и концентрации анализируемых веществ.
concentrations = pd.read_csv('http://stepik.org/media/attachments/course/4852/algae.csv')
```

```python
concentrations
```

```python
#1. Помогите Ростиславу найти среднюю концентрацию каждого из веществ в каждом из родов (колонка genus)! 
#Для этого проведите группировку датафрэйма, сохранённого в переменной concentrations, 
#и примените метод, сохранив результат в переменной mean_concentrations.
concentrations.groupby('genus') \
    .aggregate({'sucrose' : 'mean', 'alanin' : 'mean', 'citrate' : 'mean', 'glucose' : 'mean', 'oleic_acid' : 'mean'})
```

```python
#2. Пользуясь предыдущими данными, укажите через пробел (без запятых) чему равны минимальная, средняя и максимальная 
#концентрации аланина (alanin) среди видов рода Fucus. Округлите до 2-ого знака, десятичным разделителем является точка.
#Выбор видов рода Fucus
con_Fucus = concentrations.query("genus == 'Fucus'")
```

```python
con_Fucus
```

```python
#Расчет минимального среднего и максимального значений
min_v = round(con_Fucus['alanin'].min(), 2)
mean_v = round(con_Fucus['alanin'].mean(), 2)
max_v = round(con_Fucus['alanin'].max(), 2)
print(min_v, mean_v, max_v)
```

```python
#3. Сгруппируйте данные по переменной group и соотнесите вопросы с ответами
concentrations.groupby('group').head()
```

```python
#Число видов бурых, зелёных и красных водорослей
concentrations.groupby('group').aggregate({'species': 'count'})
```

```python
#Размах (максимум - минимум) сахарозы в бурых, зелёных и красных водорослях
concentrations.groupby('group').aggregate({'sucrose': 'max'}) -concentrations.groupby('group').aggregate({'sucrose': 'min'})
```

```python
#Дисперсия концентрации цитрата в бурых, зелёных и красных водорослях
concentrations.groupby('group').aggregate({'citrate': 'var'})
```
