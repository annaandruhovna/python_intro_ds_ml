```python
import pandas as pd
import numpy as np
```

```python
students_perfomance = pd.read_csv('https://stepik.org/media/attachments/course/4852/StudentsPerformance.csv')
```

```python
#Переименование переменных в датафрейме
students_perfomance = students_perfomance \
    .rename(columns = 
            {'parental level of education' : 'parental_level_of_education',
             'math score' : 'math_score',
             'reading score' : 'reading_score',
             'writing score' : 'writing_score'
            })
```

```python
students_perfomance.head()
```

```python
#Применение groupby вместе с aggregate
students_perfomance.groupby('gender').aggregate({'math_score' : 'mean', 'reading_score' : 'mean'})
```

```python
#Применение groupby вместе с aggregate с выделением отдельной колонки для гендера
students_perfomance.groupby('gender', as_index=False)\
    .aggregate({'math_score' : 'mean', 'reading_score' : 'mean'}) \
    .rename(columns = {'math_score' : 'mean_math_score', 'reading_score' : 'reading_score'})
```

```python
#Применение groupby по двум переменным вместе с aggregate с выделением отдельных колонок для этих переменых
students_perfomance.groupby(['gender','race/ethnicity'], as_index=False)\
    .aggregate({'math_score' : 'mean', 'reading_score' : 'mean'}) \
    .rename(columns = {'math_score' : 'mean_math_score', 'reading_score' : 'reading_score'})
```

```python
#Применение groupby по двум переменным вместе с aggregate с мультииндексом
mean_scores = students_perfomance.groupby(['gender','race/ethnicity'])\
    .aggregate({'math_score' : 'mean', 'reading_score' : 'mean'}) \
    .rename(columns = {'math_score' : 'mean_math_score', 'reading_score' : 'reading_score'})
```

```python
mean_scores
```

```python
#Обращение к данным датафрейма черех мультииндекс
mean_scores.loc[[('female', 'group A'), ('female', 'group B')]]
```

```python
#Сортировка значений по убыванию по гендеру и баллу по математике 
students_perfomance.sort_values(['gender','math_score'], ascending=False) \
    .groupby('gender').head(5)
```
