```python
import pandas as pd
import numpy as np
```

```python
students_perfomance = pd.read_csv('https://stepik.org/media/attachments/course/4852/StudentsPerformance.csv')
```

```python
students_perfomance
```

```python
#Список переменных
list(students_perfomance)
```

```python
#Переименование переменных в датафрейме
students_perfomance = students_perfomance \
    .rename(columns = 
            {'parental level of education' : 'parental_level_of_education',
             'math score' : 'math_score',
             'reading score' : 'reading_score',
             'writing score' : 'writing_score'
            })
```

```python
#Выбор переменных, в которых есть в названии score
score_columns = [i for i in list(students_perfomance) if 'score' in i]
```

```python
students_perfomance[score_columns].head()
```

```python
#Выбор переменных, в которых есть в названии score через функцию фильтр
#Работает и для строк
students_perfomance.filter(like='score').head()
```
