```python
import pandas as pd
import numpy as np
```

```python
students_perfomance = pd.read_csv('https://stepik.org/media/attachments/course/4852/StudentsPerformance.csv')
```

```python
students_perfomance
```

```python
students_perfomance.gender == 'female'
```

```python
#Отбор только тех строк датафрейма,где пол женский + оставить только 2 колонки
students_perfomance.loc[students_perfomance.gender == 'female', ['gender', 'lunch']]
```

```python
#Среднее значение для серии из датафрейма
mean_writing_score = students_perfomance['writing score'].mean()
```

```python
#Отбор тех строк, где writing score выше среднего
students_perfomance.loc[students_perfomance['writing score'] > mean_writing_score]
```

```python
students_perfomance[(students_perfomance['writing score'] > 100) & (students_perfomance.gender == 'female')]
```

```python
#Отбор ячеек через несколько условий
students_perfomance[(students_perfomance['writing score'] > 50) & (students_perfomance.gender == 'female')]
```

```python
#У какой доли студентов из датасэта в колонке lunch указано free/reduced?
count_student = students_perfomance.shape[0]
count_free = students_perfomance[students_perfomance['lunch'] == 'free/reduced'].shape[0]
persent = count_free / count_student
persent
```

```python
#Как различается среднее и дисперсия оценок по предметам у групп студентов со стандартным или урезанным ланчем?
#Группировка датафрейма по параметру ланч и нахождение средних
students_perfomance.groupby('lunch').aggregate({'math score': 'mean', 'reading score': 'mean','writing score' : 'mean'})
```

```python
#Как различается среднее и дисперсия оценок по предметам у групп студентов со стандартным или урезанным ланчем?
#Группировка датафрейма по параметру ланч и нахождение дисперсий
students_perfomance.groupby('lunch').aggregate({'math score': 'var', 'reading score': 'var','writing score' : 'var'})
```
