```python
import pandas as pd
import numpy as np
```

```python
students_perfomance = pd.read_csv('https://stepik.org/media/attachments/course/4852/StudentsPerformance.csv')
```

```python
students_perfomance
```

```python
#Переименование переменных в датафрейме
students_perfomance = students_perfomance \
    .rename(columns = 
            {'parental level of education' : 'parental_level_of_education',
             'math score' : 'math_score',
             'reading score' : 'reading_score',
             'writing score' : 'writing_score'
            })
```

```python
students_perfomance
```

```python
#Получение конкретных строк через запрос query
students_perfomance.query("writing_score > 74")
```

```python
#Запрос для проверки равенства строки
students_perfomance.query("gender == 'female'")
```

```python
#Объединение двух запросов через и
students_perfomance.query("gender == 'female' & writing_score > 74")
```

```python
#Объединение двух запросов через или
students_perfomance.query("gender == 'female' | writing_score > 74")
```

```python
#Сравнение с переменной через экранирование переменной
value = 74
students_perfomance.query("writing_score > @value")
```
