```python
import pandas as pd
import numpy as np
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
students_perfomance = pd.read_csv('https://stepik.org/media/attachments/course/4852/StudentsPerformance.csv')
```

```python
students_perfomance = students_perfomance \
    .rename(columns = 
            {'parental level of education' : 'parental_level_of_education',
             'math score' : 'math_score',
             'reading score' : 'reading_score',
             'writing score' : 'writing_score'
            })
```

```python
students_perfomance.head()
```

```python
students_perfomance.math_score.hist()
```

```python
students_perfomance.math_score.plot()
```

```python
students_perfomance.plot.scatter(x='math_score', y='reading_score')
```

```python
sns.lmplot(x='math_score', y='reading_score', data=students_perfomance)
```

```python
ax=sns.lmplot(x='math_score', y='reading_score', hue='gender', fit_reg=False,data=students_perfomance)
ax.set_xlabels('Math score')
ax.set_ylabels('Reading score')
```
